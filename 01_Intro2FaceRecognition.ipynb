### <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">لود کتابخانه‌های مورد استفاده</div>

```python
import FaceToolKit as ftk
import matplotlib.pyplot as plt
import numpy as np
```

### <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">مشخص کردن  حد آستانه واندازه ورودی شبکه</div>

```python
verification_threshhold = 1.188
image_size = 160
```

### <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">ایجاد شی از کلاس نوشته شده</div>

```python
v = ftk.Verification()
```

### <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">لود یک مدل آموزش داده شده و مشخص کردن تنسورهای ورودی و خروجی شبکه</div>

```python
# Pre-load model for Verification
v.load_model("./models/20180204-160909/")
v.initial_input_output_tensors()
```

### <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">لود تصاویر از دیسک</div>



```python
#read images
img1 = plt.imread("./images/1.jpg")
img2 = plt.imread("./images/2.jpg")
img3 = plt.imread("./images/3.jpg")
```

```python
# Display the images
plt.imshow(img1)
plt.show()
plt.imshow(img2)
plt.show()
plt.imshow(img3)
plt.show()
```

### <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">محاسبه‌ی embedding ها</div>



```python
#generate embeddings
emb1 = v.img_to_encoding(img1, image_size)
emb2 = v.img_to_encoding(img2, image_size)
emb3 = v.img_to_encoding(img3, image_size)
```

```python
emb1.shape
```

### <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">تابع برای محاسبه فاصله دو بردار</div>

```python
def distance(emb1, emb2):
    diff = np.subtract(emb1, emb2)
    return np.sum(np.square(diff))
```

## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">تطبیق چهره</div>



```python
#distance
dist = distance(emb1, emb2)
is_same = dist < verification_threshhold
print ("distance img1 and img2 =", dist, " issame =", is_same)

dist = distance(emb1, emb3)
is_same = dist < verification_threshhold
print ("distance img1 and img3 =", dist, " issame =", is_same)
```
